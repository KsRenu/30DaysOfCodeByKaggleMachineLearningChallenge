# GROUPING AND SORTING

### GROUPING ANALYSIS

```python
#grouping the data and performing some actions
weather.groupby('Temperature').Temperature.count()
```

```python
#value_counts is just a to groupby()
weather.groupby('Temperature').Temperature.min()
```

```python
#generating simple statistical summary of df using agg()
weather.groupby(['Temperature']).Temperature.agg([len,min,max])
```

### MULTI-INDEXES

```python
# #differs from index by having multiple value
climate=weather.groupby(['Temperature','Humidity'].agg([len]))
climate
```

```python
type(MI)=climate.index
#it prints the output as multi index
```

```python
#to convert back to regular class
climate.reset_index()
```

## SORTING

```python
#grouping returns datas in the order of index bot by data.
#to sort by datas
climate=climate.sort_values(by='len')

#to sort in descending order
climate=climate.sort_values(by='len',ascending=False)

#to sort by index
climate.sort_index()

#to sort more than one column
climate.sort_values(by=['Temperature','Humidity'])

```
